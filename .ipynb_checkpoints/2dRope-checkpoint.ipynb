```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="7"
import numpy as np
import scipy
import tensorflow as tf
from IPython.display import clear_output
import plotly.express as px
import pandas as pd
import time
import math
import plotly.graph_objects as go
cos = math.cos
sin = math.sin
pi = math.pi
exp = tf.math.exp
```

```python
n = 100
d = 2*pi/d
```

```python
X = []
for i in range(n):
    X.append(tf.Variable([cos(2*pi*i/n), sin(2*pi*i/n)], dtype='float'))
X.append(X[0])
df = df_empty = pd.DataFrame(columns=['x', 'y'])
for i in range(n):
    df.loc[i] = X[i].numpy()
def rho(i, j):
    return min([abs(i-j), 100-abs(i-j)])
def angle(a, b):
    return sum(a * b) / tf.norm(a) / tf.norm(n) 
```

```python
def loss():
    surface = abs(sum([X[i][1]*X[i+1][0] - X[i][0]*X[i+1][1] for i in range(n)]))
    circumference = sum([ tf.norm(X[i]-X[i+1]) for i in range(n) ])
    interval = sum([ ( tf.norm(X[i]-X[i+1]) - 2*pi/d ) ** 6 for i in range(n) ])
    repulsive = tf.constant(0.0)
    for i in range(n):
        for j in range(n):
            if i<j:
                repulsive = repulsive + rho(i,j)*exp(-tf.nn.l2_loss(X[i]-X[j])/d**2)
    central = sum([tf.nn.l2_loss(X[i]) for i in range(n)])
    flexural = sum([angle(X[i]-X) for i in range(n)])
    return repulsive + central
```

```python
opt = tf.keras.optimizers.SGD(learning_rate=0.01)
fig = go.Figure(go.Scatter(x=df['x'], y=df['y'], mode='markers', 
                           marker=dict(color=df.index)))
fig.update_layout(width=500, height=500, title='%f'%loss())
                
for epoch in range(100):
    clear_output(wait=True)
    opt.minimize(loss, var_list=X[0:n])
    for i in range(n):
        df.loc[i] = X[i].numpy()
    #fig = px.scatter(df, x='x', y='y', color=df.index, title='%f'%loss())
    fig.update_traces(x=df['x'], y=df['y'])
    fig.update_layout(title='%f'%loss())
    fig.show()
    time.sleep(0.01)
```
